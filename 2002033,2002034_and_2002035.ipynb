```python

import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('/content/Raw Data.xlsx')
```

```python
data['Gender'].replace({"Male":0,"Female":1},inplace=True)
```

```python
x=data.iloc[:,1:4].values
y=data.iloc[:, -1].values
```

```python
print(x)
print(y)
```

```python
print(x.shape)
print(y.shape)
```

```python
clx1=DecisionTreeClassifier(max_depth=4)
clx1.fit(x,y)
```

# New Section

```python
from sklearn.tree import plot_tree
plot_tree(clx1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clx1=DecisionTreeClassifier()
```

```python
clx1=clx1.fit(x_train, y_train)
print(x_train.shape)
print(y_train.shape)
```

```python
y_pred=clx1.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
param_dist={
    "criterion":["giny","entropy"],
    "max_depth":[1,2,3,4,5,6,7,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(clx1,param_dist,cv=10,n_jobs=-1)
```

```python
grid.fit(x_train, y_train)
```
